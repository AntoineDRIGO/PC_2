# I have split the excel into several tables: Country, HV groups, Regions and A

```python
import pandas as pd
#I start with the table "Country"
url = 'https://github.com/AntoineDRIGO/PC_2/raw/main/Country.csv'

# I read csv precising "," is the way to get decimals
country = pd.read_csv(url, sep=';', decimal=',', encoding='ISO-8859-1')
country
```

```python
# I want to delate the 2 last empty columns
country = country.iloc[:, :9]
country
```

```python
#I rename the columns
country = country.rename(columns={
    'HDI rank': 'HDI_rank_2021','Human Development Index (HDI)':'HDI','Life expectancy at birth':'Life_expectancy',
    'Expected years of schooling': 'School_level_expected','Mean years of schooling':'Mean_school_level',
    'Gross national income (GNI) per capita':'GNI','GNI per capita rank minus HDI rank': 'GNI_rank','HDI rank 2020':'HDI_rank_2020'})
country
```

```python
print(country.dtypes)
```

```python
#I want to get the GNI_rank column a float to use it like a number
country.GNI_rank = country.GNI_rank.astype(float)
type(country.GNI_rank[0])
```

```python
# I delate the spaces in the GNI column to transform it in float
country.GNI = country.GNI.str.replace(' ', '')
country.GNI = country.GNI.astype(float)
type(country.GNI[0])
```

```python
# Find last index where there is no values in the Country column
last_valid_index = country['Country'].last_valid_index()

# Delate all the lines after
country = country.loc[:last_valid_index]
country
```
